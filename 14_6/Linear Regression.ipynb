## Normalise the dataset

## q={q0,q1}

```python
theta=np.zeros((2,1))
print(theta)
```

```python
def grad(X,Y,theta):
    m=X.shape[0]
    grad=np.zeros((2,))
    for i in range(m):
        yi=theta[1]*X[i] + theta[0]
        grad[0]+=2*(yi-Y[i])
        grad[1]+=2*(yi-Y[i])*X[i]
        return grad
            
            
```

```python
def total_error(X,Y,theta):
    m=X.shape[0]
    e=0
    for i in range(m):
        yi=theta[1]*X[i]+theta[0]
        e+=(yi-Y[i])**2
    return e
            
```

```python
def line(X,q0,q1):
    return q1*X+q0
```

```python
def gradientDescent(X,Y,learning_rate=0.00001,max_steps=1980):
    theta=np.zeros((2,))
    error_list=[]
    for i in range(max_steps):
        #update theta
        g=grad(X,Y,theta)
        theta[0]=theta[0]-learning_rate*g[0]
        theta[1]=theta[1]-learning_rate*g[1]
        #plt.plot(X,line(X,theta[0],theta[1]))
        error=total_error(X,Y,theta)
        error_list.append(error)
    #plt.show()
    return theta,error_list
```

```python
plt.figure(figsize=(13,13))
plt.scatter(X,Y)
plt.scatter(X,line(X,theta[0],theta[1]))
theta,error_list=gradientDescent(X,Y)
```
