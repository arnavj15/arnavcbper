```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
print(x)
```

```python
y=(x-5)**2
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
## gradient Descent
```

```python
steps=1
x1=10
learning_rate=0.1
plt.plot(x,y)

while(steps<=30):
    x1=x1-learning_rate*2*(x1-5)
    error=((x1-5)**2)
    plt.scatter(x1,error)
    print("X %.4f,Error %.4f"%(x1,error))
    steps+=1
plt.show()
```
