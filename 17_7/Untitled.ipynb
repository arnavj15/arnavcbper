```python
def genTransitionTable(text):
    T={}
    k=4
    for i in range(len(text)-k):
        X=text[i:i+k]
        Y=text[i+k]
        if T.get(X) is None:
            T[X]={}
            T[X][Y]=1
        else:
            if T[X].get(Y) is None:
                T[X][Y]=1
            else:
                T[X][Y]+=1
            
    #NOrmalize
    
    for kx in T.keys():
        s=0
        for v in T[kx].values():
            s+=v
        for ky in T[kx].keys():
            T[kx][ky]=T[kx][ky]/s
            
    return T
    
    
    
```
