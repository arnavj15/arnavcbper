```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.arange(10)
x2=x**2
```

```python
plt.figure(figsize=(3,3))
plt.plot(x,label="linear")
plt.plot(x2,label="Quadratic")
plt.legend()
plt.xlabel("n")
plt.show()
```

```python
plt.scatter(x,x2,color="orange")
```
